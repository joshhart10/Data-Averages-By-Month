```python
### Importing Data Functions ###

# This function is for when the user inputs a station number to find the streamflow data
def stationNumber_function():

    # Find the USGS web link matching the user's inputs
    section1 = 'https://nwis.waterdata.usgs.gov/nwis/dv?referred_module=sw&search_site_no='
    section2 = '&search_site_no_match_type=exact&site_tp_cd=OC&site_tp_cd=OC-CO&site_tp_cd=ES&site_tp_cd='\
    'LK&site_tp_cd=ST&site_tp_cd=ST-CA&site_tp_cd=ST-DCH&site_tp_cd=ST-TS&index_pmcode_00060=1&group_key='\
    'NONE&sitefile_output_format=html_table&column_name=agency_cd&column_name=site_no&column_name=station_nm&range_selection=date_range&begin_date='
    section3 = '&end_date='
    section4 = '&format=rdb&date_format=YYYY-MM-DD&rdb_compression=value&list_of_search_criteria=search_site_no%2Csite_tp_cd%2Crealtime_parameter_selection'

    link = (section1 + station_number + section2 + start_date + section3 + end_date + section4)
    print("Click here to see the generated USGS link: \n", link)

    # Read the webpage and put data into a variable
    USGS_page = urllib.request.urlopen(link)
    downloaded_data = USGS_page.read()

    # Format data as a string instead of bytes
    str_data = downloaded_data.decode()

    # Seperate the data by each new line. Every line of data will be split into it's own index
    f_str_data = str_data.split('\n')

    # Find station name from data and make it a global variable
    global station_name
    station_name = ''
    # Loop through every line in the data and look for the line that contains the station name
    for line in range(len(f_str_data)):
        if f_str_data[line].startswith("#    USGS"):
            # Drop the the "#"" and the "USGS" from the station name
            station_name = f_str_data[line][5:]

    # Get rid of all columns that aren't date and streamflow
    date_flow = ''

    # Loop through all lines of data in the table
    for line in range(len(f_str_data)):
        if f_str_data[line].startswith("USGS"):
            # Drop the first part of the string leaving only the characters after the date
            data = f_str_data[line][14:]
            # Split the data by where it is tabbed seperating it into columns of date, streamflow and publication
            columns = data.split('\t')
            # Drop the thrid column
            rows = ','.join([columns[0],(columns[1])])
            date_flow += rows + '\n'
    date_flow = date_flow.encode()

    # Create a file from the data and upload it to the runtime
    folder = os.getcwd()
    with open(folder + '/USGS_Data_for_' + station_number  + '.txt', 'wb') as text:
        text.write(date_flow)
    filename = folder + '/USGS_Data_for_' + station_number  + '.txt'

    # Name the columns in the dataframe that's about to be created
    columns = ['Date','Streamflow (cfs)']

    # Turn the data into a dataframe using your file, and new column names
    # parse_dates = [0] is turning the first column into datetime objects
    global streamflowData_df
    streamflowData_df = pd.read_csv(filename, header = None, names = columns, parse_dates = [0])

def csv_function():
    # Prompt user to upload csv file with dates and streamflow data
    from google.colab import files

    uploaded = files.upload()

    # Program gets the path to your CSV file
    csv_file_path = list(uploaded.keys())[0]

    # Getting station name for plots down below
    global station_name
    station_name = csv_file_path

    # Read the CSV file into a DataFrame
    global streamflowData_df
    streamflowData_df = pd.read_csv(csv_file_path, header = None)

    # Add column names to the dataframe
    streamflowData_df.columns = ['Date', 'Streamflow (cfs)']


### User Input ###

# Give user two options of uploading a csv file or inputting a USGS station number
station_number = input("(Type 'csv' if you wish to import a csv instead)\nWhat is the USGS Station ID?\t")
if station_number.lower() == 'csv':
    csv_function()
else:
    start_date = input("Start Date (YYYY-MM-DD):\t")
    end_date = input("End Date (YYYY-MM-DD):\t\t")
    stationNumber_function()
```

```python
### Average Streamflow by Month ###


jan_indices = list(streamflowData_df.loc[streamflowData_df['Month'] == 1].index)
feb_indices = list(streamflowData_df.loc[streamflowData_df['Month'] == 2].index)
mar_indices = list(streamflowData_df.loc[streamflowData_df['Month'] == 3].index)
apr_indices = list(streamflowData_df.loc[streamflowData_df['Month'] == 4].index)
may_indices = list(streamflowData_df.loc[streamflowData_df['Month'] == 5].index)
jun_indices = list(streamflowData_df.loc[streamflowData_df['Month'] == 6].index)
jul_indices = list(streamflowData_df.loc[streamflowData_df['Month'] == 7].index)
aug_indices = list(streamflowData_df.loc[streamflowData_df['Month'] == 8].index)
sep_indices = list(streamflowData_df.loc[streamflowData_df['Month'] == 9].index)
oct_indices = list(streamflowData_df.loc[streamflowData_df['Month'] == 10].index)
nov_indices = list(streamflowData_df.loc[streamflowData_df['Month'] == 11].index)
dec_indices = list(streamflowData_df.loc[streamflowData_df['Month'] == 12].index)



streamflowData_df.loc[jan_indices,'Monthly Averages'] = streamflowData_df['Streamflow (cfs)'][jan_indices].mean()
streamflowData_df.loc[feb_indices,'Monthly Averages'] = streamflowData_df['Streamflow (cfs)'][feb_indices].mean()
streamflowData_df.loc[mar_indices,'Monthly Averages'] = streamflowData_df['Streamflow (cfs)'][mar_indices].mean()
streamflowData_df.loc[apr_indices,'Monthly Averages'] = streamflowData_df['Streamflow (cfs)'][apr_indices].mean()
streamflowData_df.loc[may_indices,'Monthly Averages'] = streamflowData_df['Streamflow (cfs)'][may_indices].mean()
streamflowData_df.loc[jun_indices,'Monthly Averages'] = streamflowData_df['Streamflow (cfs)'][jun_indices].mean()
streamflowData_df.loc[jul_indices,'Monthly Averages'] = streamflowData_df['Streamflow (cfs)'][jul_indices].mean()
streamflowData_df.loc[aug_indices,'Monthly Averages'] = streamflowData_df['Streamflow (cfs)'][aug_indices].mean()
streamflowData_df.loc[sep_indices,'Monthly Averages'] = streamflowData_df['Streamflow (cfs)'][sep_indices].mean()
streamflowData_df.loc[oct_indices,'Monthly Averages'] = streamflowData_df['Streamflow (cfs)'][oct_indices].mean()
streamflowData_df.loc[nov_indices,'Monthly Averages'] = streamflowData_df['Streamflow (cfs)'][nov_indices].mean()
streamflowData_df.loc[dec_indices,'Monthly Averages'] = streamflowData_df['Streamflow (cfs)'][dec_indices].mean()


# Plot monthly flow
plt.plot(streamflowData_df['Month'][0:365], streamflowData_df['Monthly Averages'][0:365])

plt.xlabel('Month')
plt.xticks(list(range(1,13)))
plt.ylabel('Average Streamflow (cfs)')
plt.title('Average Streamflow by Month from ' + str(first_date) + ' to ' + str(last_date))
plt.show()
```
